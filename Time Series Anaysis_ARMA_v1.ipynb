Generate Samples From ARMA(3,4)

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
np.random.seed(12345)
arparams = np.array([.75, -.25])
maparams = np.array([.65, .35])
ar = np.r_[1, -arparams] 
ma = np.r_[1, maparams] 
y = sm.tsa.arma_generate_sample(ar, ma, 100)
plt.plot(y, linestyle='-', color='b')
plt.title('Simulated ARMA(3,4) Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
