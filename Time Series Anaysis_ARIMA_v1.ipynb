```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

np.random.seed(42)

n_obs = 100
t = np.arange(n_obs)
data = 2 * np.sin(0.2 * t) + np.random.normal(0, 1, n_obs)

data = np.cumsum(data)

order = (4, 3, 2)
model = sm.tsa.ARIMA(data, order=order)
results = model.fit()

print(results.summary())

plt.figure(figsize=(10, 6))
plt.plot(t, data, label='Original Data', color='b')
plt.plot(t, results.fittedvalues, label='Fitted Values', color='r', linestyle='--')
plt.title('ARIMA(4, 3, 2) Model Fit')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
